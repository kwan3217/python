# Historical Background

The year Tycho Brahe died (1601) Johannes Kepler, who had worked with Tycho in the 18 months preceding his death, was appointed as Imperial Mathematician to the Court of Emperor Rudolph II. Recognizing the goldmine of information locked up in Tycho’s painstaking observations, Kepler packed them up and moved them to Prague with him. In a letter to one of his English admirers he calmly reported: 

>"I confess that when Tycho died, I quickly took advantage of the absence, or lack of circumspection, of the heirs, by taking the observations under my care, or perhaps usurping them .. . "

Kepler stayed in Prague from 1601 to 1612. It was the most fruitful period of his life and saw the publication of *Astronomia Nova* in 1609 in which he announced his first two laws of planetary motion. The manner in which Kepler arrived at these two laws is fascinating and can be told only because in *New Astronomy* Kepler leads his reader into every blind alley, detour, trap or pitfall that he himself encountered. 

> “What matters to me,” Kepler points out in his Preface, "is not merely to impart to the reader what I have to say, but above all to convey to him the reasons, subterfuges, and lucky hazards which led me to my discoveries. When Christopher Columbus, Magelhaen and the Portuguese relate how they went astray on their journeys, we not only forgive them, but would regret to miss their narration because without it the whole grand entertainment would be lost."

Kepler selected Tycho’s observations of Mars and tried to reconcile them with some simple geometrical theory of motion. He began by making three revolutionary assumptions: (a) that the orbit was a circle with the sun slightly off-center, (b) that the orbital motion took place in a plane which was fixed in space, and (c) that Mars did not necessarily move with uniform velocity along this circle. Thus, Kepler immediately cleared away a vast amount of rubbish that had obstructed progress since Ptolemy. 

Kepler’s first task was to determine the radius of the circle and the

At the very beginning of a whole chapter of excruciating trial-and-error calculations, Kepler absentmindedly put down three erroneous figures for three vital longitudes of Mars, never noticing his error. His results, however, were nearly correct because of several mistakes of simple arithmetic committed later in the chapter which happened very nearly to cancel out his earlier errors. 

At the end he seemed to have achieved his goal of representing within 2 arc-minutes the position of Mars at all 10 oppositions recorded by Tycho. But then without a word of transition, in the next two chapters Kepler explains, almost with masochistic delight, how two other observations from Tycho’s collection did not fit; there was a

Before Kepler could determine the true shape of Mars’ orbit, without benefit of any preconceived notions, he had to determine precisely the earth’s motion around the sun. For this purpose he designed a highly original method and when he had finished his computations he was ecstatic. At this point Kepler could contain himself no longer and becomes airborne, as it were, with the warning: "Ye physicists, prick your ears, for now we are going to invade your territory." This was the first of the critical mistakes that would cancel itself out "as if by a miracle" and lead him by faulty reasoning to the correct result. 

Forgetting his earlier resolve to abandon circular motion he reasoned, again incorrectly, that, since speed was inversely proportional to distance, the line joining the sun (which was off-center in the circle) and the planet swept out equal areas in the orbit in equal times.

# Time-of-Flight as a Function of Eccentric Anomaly
Many of the concepts introduced by Kepler, along with the names he used to describe them, have persisted to this day. You are already familiar with the term "true anomaly" used to describe the angle from periapsis to the orbiting object measured in the direction of motion. In this section we will encounter a new term called "eccentric anomaly" which was introduced by Kepler in connection with elliptical orbits. Although he was not aware that parabolic and hyperbolic orbits existed, the concept can be extended to these orbits also as we shall see. 

It is possible to derive time-of-flight equations analytically, using only the dynamical equation of motion and integral calculus. We will pursue a lengthier, but more motivated, derivation in which the eccentric anomaly arises quite naturally in the course of the geometrical arguments. This derivation is presented more for its historical value than for actual use. The universal variable approach is strongly recommended as the best method for general use.

## Time-of-Flight on the Elliptical Orbit
We have already seen that in one orbital period the radius vector sweeps out an area equal to the total area of an ellipse, i.e., $\pi ab$. In going part way around an orbit, say from periapsis to some general point, P, where the true anomaly is $\nu$, the radius vector sweeps out the shaded area, $A_1$, in Figure 4.2-1. Because area is swept out at a constant rate in an orbit (Kepler’s Second Law) we can say that 

$$\frac{t-T}{A_1}=\frac{\mathbb{P}}{\pi ab}\tag{4.2-1}$$

where $T$ is the time of periapsis passage and $\mathbb{P}$ is the period. 
>*Margin Note: The term on the right is the reciprocal of the areal rate. Since the radius vector sweeps out equal areas in equal times, it sweeps out area at a constant time rate. The term on the right thus is the period of the whole orbit divided by the whole area of the whole ellipse. Note that while we could figure out what $\mathbb{P}$ is, since it's calculable from Kepler's third law, we don't care what it's value is yet. The term on the left is the time since periapse divided by the area swept out since periapse.*


Figure 4.2-1 Area swept out by $\vec{r}$ 


The only unknown in equation (4.2-1) is the area $A_1$. The geometrical construction illustrated in Figure 4.2-2 will enable us to write an expression for $A_1$. 
>*We might be able to integrate this area, but Kepler didn't have integration. The figure below has that equation $E=\frac{\mbox{area }QOV}{1/2a^2}$ without derivation. The area of a whole circle is (don't say tau, don't say tau...) $\frac{1}{2}2\pi r^2$, and a sector of angle $\theta$ (in radaians) is $\frac{\theta}{2\pi}$ of the area of a whole circle since a whole circle is $2\pi$ radians. Therefore the area of a sector is: $$\begin{eqnarray*}
A_{\theta}&=&\frac{\theta}{2\pi}1/2(2\pi)r^2\\
A_{\theta}&=&\theta(1/2)r^2\\
\end{eqnarray*}$$ In our case, $E=\theta$, $A_\theta=\mbox{area }QOV$, and $a=r$, so: $$\begin{eqnarray*}
\mbox{area }QOV&=&E(1/2)a^2 \\
E&=&\frac{\mbox{area }QOV}{1/2a^2}
\end{eqnarray*}$$*


Figure 4.2-2 Eccentric anomaly, $E$

A circle of radius $a$ has been circumscribed about the ellipse. A dotted line, perpendicular to the major axis, has been extended through $P$ to where it intersects the "auxiliary circle" at $Q$. The angle $E$ is called the *eccentric anomaly*. 

Before proceeding further we must derive a simple relationship between the ellipse and its auxiliary circle. From analytical geometry, the equations of the curves in cartesian coordinates are: 

$$\mbox{Ellipse: }\frac{x^2}{a^2}+\frac{y^2}{b^2}=1\,\,\,\mbox{Circle: }\frac{x^2}{a^2}+\frac{y^2}{a^2}=1$$ 

From which 

$$y_{ellipse}=\sqrt{\frac{a^2b^2-b^2x^2}{a^2}}=\frac{b}{a}\sqrt{a^2-x^2}$$
$$y_{circle}=\sqrt{a^2-x^2}$$

Hence 

$$\frac{y_{ellipse}}{y_{circle}}=\frac{b}{a}\tag{4.2-2}$$

This simple relationship between the $y$-ordinates of the two curves will play a key role in subsequent area and length comparisons. 

From Figure 4.2-2 we note that the area swept out by the radius vector is Area $PSV$ minus the dotted area, $A_2$. 

$$A_1=\mbox{Area }PSV-A_2.$$

Since $A_2$ is the area of a triangle whose base is $ae - a \cos E$ and whose altitude is $b/a (a \sin E)$, we can write 

$$A_2 =\frac{ab}{2}(e\sin E - \cos E \sin E).\tag{4.2-3}$$

Area $PSV$ is the area under the ellipse; it is bounded by the dotted line and the major axis. Area $QSV$ is the corresponding area under the auxiliary circle. It follows directly from equation (4.2-2) that 

$$\mbox{Area }PSV=\frac{b}{a}(\mbox{Area }QSV)$$


Figure 4.2-3 Area QSV 

The Area $QSV$ is just the area of the sector $QOV$, which is $1/2 a^2 E$ (where $E$ is in radians), minus the triangle, whose base is $(a \cos E)$, and whose altitude is $(a \sin E)$. Hence 

$$\mbox{Area }PSV = \frac{ab}{2}(E - \cos E\sin E).$$

Substituting into the expression for area $A_1$ yields 

$$A_1 = \frac{ab}{2}(E - e \sin E).$$

Finally, substituting into equation (4.2-1) and expressing the period as $2\pi\sqrt{a^3/\mu}$,, we get

>$$t-T=\sqrt{\frac{a^3}{\mu}}(E-e\sin E)\tag{4.2-4}$$

Kepler introduced the definition 

$$M = E - e \sin E\tag{4.2-5}$$

where M is called the "mean anomaly." If we also use the definition 

$$n\equiv \sqrt{\mu/a^3}$$

where $n$ is called the "mean motion," then the mean anomaly may be written: 

$$M = n(t - T) = E - e \sin E\tag{4.2-6}$$

which is often referred to as *Kepler’s equation*. 

Obviously, in order to use equation (4.2-4), we must be able to relate the eccentric anomaly, $E$, to its corresponding true anomaly, $\nu$. From Figure 4.2-2, 

$$\cos E = \frac{ae+r\cos \nu}{a}.\tag{4.2-7}$$ 

Since $r=\frac{a(1-e^2)}{1+e\cos\nu} equation (4.2-7) reduces to 

$$\cos E=\frac{e+\cos\nu}{1+e\cos\nu}.\tag{4.2-8}$$

Eccentric anomaly may be determined from equation (4.2-8). The correct quadrant for $E$ is obtained by noting that $\nu$ and $E$ are always in the same half-plane; when $\nu$ is between 0 and $\pi$, so is $E$. 


Figure 4.2-4 Time-of-flight between arbitrary points 

>*This next section is so obvious that even I wouldn't have done it explicitly. Once you have the time betweeen a reference point (here periapse) and any arbitrary points, you can figure the time between any two arbitrary points by considering the time between each arbitrary point and the reference point*

Suppose we want to find the time-of-flight between a point defined by $\nu_0$ and some general point defined by $\nu$ when the initial point is not periapsis. Provided the object does not pass through periapsis enroute from $\nu_0$ to $\nu$ (Figure 4.2-4 left), we can say that 

$$t-t_0=(t-T)-(t_0-T).$$

If the object does pass through periapsis (which is the case whenever $\nu_0$ is greater than $\nu$) then, from Figure 4.2-4 right, 

$$t-t_0 = \mathbb{P}+(t-T)-(t_0-T_0).$$

In general we can say that
>$$\begin{eqnarray*}
t-t_0&=&\sqrt{\frac{a^3}{\mu}}[2k\pi+(E-e\sin E) \\
     & &-(E_0-e\sin E_0)]
\end{eqnarray*}\tag{4.2-9}$$

where $k$ is the number of times the object passes through periapsis enroute from $\nu_0$ to $\nu$. 

At this point it is instructive to note that this same result can be derived analytically. In this case the eocentric anomaly appears as a convenient variable transformation to permit integration. Only the skeleton of the derivation will be shown here. 

From equations in section 1.7.2 we can write 

$$\int_T^t hdt=\int_0^\nu r^2d\nu\tag{4.2-10}$$

or

$$h(t-T)=\int_0^\nu\frac{p^2d\nu}{(1+e\cos \nu)^2}\tag{4.2-11}$$

Now let us introduce the eccentric anomaly as a variable change to make equation (4.2-11) easily integrable. From equation (4.2-8) and geometry the following relationships can be derived: 

$$\begin{eqnarray*}
\cos\nu&=&\frac{e-\cos E}{e \cos E-1}\tag{4.2-12} \\
\sin\nu&=&\frac{a\sqrt{1-e^2}}{r}\sin E\tag{4.2-13} \\
r&=&a(1-e\cos E)\tag{4.2-14}
\end{eqnarray*}$$

Differentiating equation (4.2-12), we obtain 

$$\begin{eqnarray*}
d\nu&=&\frac{\sin E(1+e\cos\nu)}{\sin\nu(1-e\cos E)}dE=\frac{\sin E(p/r)}{\sin\nu(r/a)} \\
 &=&\frac{a\sqrt{1-e^2}}{r}dE.\tag{4.2-15}
\end{eqnarray*}$$

then 

$$\begin{eqnarray*}
h(t-T)&=&\frac{p}{\sqrt{1-e^2}}\int_0^Er\,dE \\
 &=&\frac{pa}{\sqrt{1-e^2}}\int_0^E (1-e\cos E)\,dE \\
 &=&\frac{pa}{\sqrt{1-e^2}}(E-e\sin E).\tag{4.2-16}
\end{eqnarray*}$$

since $h=\sqrt{\mu p}$

$$(t-T)=\sqrt{\frac{a^3}{\mu}}(E-e\sin E)\tag{4.2-4}$$

which is identical to the geometrical result. 

## Time-of-Flight on Parabolic and Hyperbolic Orbits. 
In a similar manner, the analytical derivation of the parabolic time-of-flight can be shown to be
>$$t-T=\frac{1}{2\sqrt{\mu}}\left[pD+\frac{1}{3}D^3\right]\tag{4.2-17}$$

or

>$$t-t_0=\frac{1}{2\sqrt{\mu}}\left[\left(pD+\frac{1}{3}D^3\right)-\left(pD_0+\frac{1}{3}D_0^3\right)\right]\tag{4.2-18}$$

4.2.3 Loss of Numerical Accuracy for Near-Parabolic Orbits. The 

Kepler time-of-flight equations suffer from a severe loss in computational accuracy near $e = 1$. The nature of the difficulty can best be illustrated by a numerical example: 

Suppose we want to compute the time-of-flight from periapsis to a point where $\nu=60^\circ$ on an elliptical orbit with $a = 100\mbox{ DU}$ and $e = 0.999$. The first step is to compute the eccentric anomaly. From equation (4.2-8), 

$$\cos E=\frac{.999+.5}{1+.999(.5)}=.99967$$

Therefore $E = 0.02559$, $\sin E = 0.02560$ and $(e sin E) = 0.02557$. 

>*Of course in the modern days of double precision, we lose just as much but start with more precision to begin with, so the answer might still be usable.*

```python
nu=np.radians(60)
a=100
e=0.999
cosE=(e+np.cos(nu))/(1+e*np.cos(nu))
print("cosE: %.5f"%cosE)
E=np.arccos(cosE)
print("E (rad): %.5f"%E)
sinE=np.sqrt(1-cosE**2)
print("sinE: %.5f"%sinE)
esinE=e*sinE
print("esinE: %.5f"%esinE)
```

>*The loss of precision is already apparent in the calculated value of $E$ and $\sin E$. For a sensitivity analysis, let's see how different things are if we are off by 1 unit in the last place:*

```python
def pcterr(calculated, accepted):
    """
    Percentage error in calculated value vs accepted value
    :param calculated: Calculated value, assumed to be in error
    :param accepted: Accepted value, assumed to be perfectly correct
    :return: percentage difference. Note not fractional difference. Positive means
             calculated value is greater than accepted.
    """
    return 100*(calculated-accepted)/accepted

cosE1=cosE+0.00001
print("cosE1: %.5f (%f%% difference)"%(cosE1,pcterr(cosE1,cosE)))
E1=np.arccos(cosE1)
print("E1: %.5f (%f%% difference)"%(E1,pcterr(E1,E)))
sinE1=np.sqrt(1-cosE1**2)
print("sinE1: %5f (%f%% difference)"%(sinE1,pcterr(sinE1,sinE)))
esinE1=e*sinE1
print("esinE1: %.5f (%f%% difference)"%(esinE1,pcterr(esinE1,esinE)))
```

Substituting these values into equation (4.2-4), 

$$\begin{eqnarray*}
t-T&=&\sqrt{\frac{100^3}{1}}(.02559 - .02557) \\
   &=&1000(.00002) = .02\mbox{ TU}
\end{eqnarray*}$$

```python
tmT=np.sqrt(a**3)*(E-esinE)
print("t-T: %f"%tmT)
tmT1=np.sqrt(a**3)*(E1-esinE1)
print("(t-t)1: %f (%f%% difference)"%(tmT1,pcterr(tmT1,tmT)))
```

There is a loss of significant digits in computing $E$ from $\cos E$ when E is near zero. There is a further loss in subtracting two nearly equal numbers in the last step. As a result, the answer is totally unreliable. 

>*In the subtraction between two nearly equal numbers, the significant figures in the difference are only those that aren't the same in the original numbers. Carrying five significant figures as we see above, only one is different, so there is only one significant figure left. In full double precision, there's still 11 digits left, but that means we have already given up 4 figures, about half the difference between single and double precision. In general, the small angle approximation is working against us in this subtraction $E-e\sin E$ under these conditions.*

This loss of computational accuracy near $e = 1$ and the inconvenience of having a different equation for each type of conic orbit will be our principal motivation for developing a universal formulation for time-of-flight in section 4.3. 

# A Universal Formulation for Time-of-Flight 

The classical formulations for time-of-flight involving the eccentric anomalies, $E$ or $F$, don't work very well for near-parabolic orbits. We have already seen the loss of numerical accuracy that can occur near $e =1$ in the Kepler equation. Also, solving for $E$ or $F$ when $a$, $e$, $\nu_0$ and $t-t_0$ are given is difficult when $e$ is nearly $1$ because the trial-and-error solutions converge too slowly or not at all. Both of these defects are 
overcome in a reformulation of the time-of-flight equations made possible by the introduction of a new auxiliary variable different from the eccentric anomaly. Furthermore, the introduction of this new auxiliary variable allows us to develop a single time-of-flight equation valid for all conic orbits. 

The change of variable is known as the "Sundman transformation" and was first proposed in 1912. Only recently, however, has it been used to develop a unified time-of-flight equation. Goodyear, Lemmon, Herrick, Stumpff, Sperling and Battin have all presented formulae for computation of time-of-flight via "generalized" or "universal" variables. The original derivation presented below was suggested by Bate and partially makes use of notation introduced by Battin. 

## Definition of the Universal Variable, $x$. 
Angular momentum and energy are related to the geometrical parameters $p$ and $a$ by the familiar equations: 

$$h=r^2\dot{\nu}=\sqrt{\mu p}$$

$$\mathcal{E}=\tfrac{1}{2}v^2-\frac{\mu}{r}=\frac{-\mu}{2a}.$$

If we resolve V into its radial component, $\dot{r}$, and its transverse component, $r\dot{\nu}$, the energy equation can be written as 

$$\tfrac{1}{2}\dot{r}^2+\tfrac{1}{2}(r\dot{\nu})^2-\frac{\mu}{r}=\frac{-\mu}{2a}.$$


Figure 4.3-1 Radial and transverse components of velocity vector, $\vec{v}$

Solving for $\dot{r}^2$ and setting $(r\dot{\nu})^2=\frac{\mu p}{r^2}$, we get 

$$\dot{r}^2=\frac{-\mu p}{r^2}+\frac{2\mu}{r}-\frac{\mu}{a}.\tag{4.3-1}$$

Since the solution of this equation is not obvious, we introduce a new independent variable, $x$, defined as 

>$$\dot{x}=\frac{\sqrt{\mu}}{r}\tag{4.3-2}$$

First we will develop a general expression for $r$ in terms of $x$. If we divide equation (4.3-1) by equation (4.3-2) squared, we obtain 

$$\left(\frac{dr}{dx}\right)^2=-p+2r-\frac{r^2}{a}.$$

Separating the variables yields 

$$dx=\frac{dr}{\sqrt{-p+2r-r^2/a}}\tag{4.3-3}$$

For $e\ne 1$ the indefinite integral, calling the constant of integration $c_0$,

$$x + c_0=\sqrt{a}\sin^{-1}\frac{(r/a-1)}{\sqrt{1-p/a}}$$

But $p=a(1-e^2)$, so $e=\sqrt{1-p/a}$ and we may write 

$$x+c_0=\sqrt{a} \sin^{-1}\frac{(r/a-1)}{e}$$

Finally, solving this equation for $r$ gives 

>$$r=a(1+e\sin\frac{x+c_0}{\sqrt{a}}).\tag{4.3-4}$$

Substituting equation (4.3-4) into the definition of the universal 
variable, equation (4.3-2), we obtain 

$$\sqrt{\mu}dt=a(1+e\sin\frac{x+c_0}{\sqrt{a})dx$$

$$\sqrt{\mu}t=ax-ae\sqrt{a}(\cos\frac{x+c_0}{\sqrt{a}}-\cos{c_0}{\sqrt{a}})\tag{4.3-5} $$

where we assumed $x = 0$ at $t = 0$. 

At this point we have developed equations for both $r$ and $t$ in terms of $x$. The constant of integration, $c_0$, has not been evaluated yet. Application of these equations will now be made to a specific problem type. 

# The Prediction Problem

With the Kepler time-of-flight equations you can easily solve for the time-of-flight, $t-t_0$, if you are given $a$, $e$, $\nu_0$ and $\nu$. The inverse problem of finding $\nu$ when you are given $a$, $e$, $\nu_0$ and $t-t_0$ is not so simple, as we shall see. Small, in *An Account of the Astronomical Discoveries of Kepler*, relates: “This problem has, ever since the time of Kepler, continued to exercise the ingenuity of the ablest geometers; but no solution of it which is rigorously accurate has been obtained. Nor is there much reason to hope that the difficulty will ever be overcome . ..” This problem classically involves the solution of Kepler’s Equation and is 
often referred to as Kepler’s problem. 

4.4.1 Development of the Universal Variable Formulation. The prediction problem can be stated as (see Figure 4.4-1): 

Given: $\vec{r}_0$, $\vec{v}_0$, $t_0=0$ 

Find: $\vec{r}$, $\vec{v}$ at time $t$. 

>*In code terms, develop a function which takes vectors $\vec{r}_0$, $\vec{v}_0$ and a time-of-flight $\Delta t$ as inputs and produces vectors $\vec{r}$ and $\vec{v}$ as outputs:*

```python
def kepler(r0v,v0v,dt):
    rv=r0v #We will replace these two lines with code that actually solves the problem
    vv=v0v
    return (rv,vv)
```

>*Notice that the input and output make no reference to the classical orbital elements $a$, $e$, etc. They are intermediate results anyway, since we can use the methods of chapter 1 to figure them out from the state vectors. Kepler himself needed these to solve his problem, but (foreshadowing) perhaps there is a way to do without using these as intermediate results.*

We have assumed $x=0$ at $t=0$. From equation (4.3-4) 

$$e\sin \frac{c_0}{\sqrt{a}}=\frac{r_0}{a}-1.\tag{4.4-1}$$

Now differentiate equation (4.3-4) with respect to time: 

$$\dot{r}=\frac{ae}{\sqrt{a}}\cos\left[\frac{(x+c_0)}{\sqrt{a}}\right]\frac{\sqrt{\mu}}{r}.\tag{4.4-2}$$

Applying the initial conditions to equation (4.4-2) and using the identity $\vec{r}\cdot\dot{\vec{r}}=r\dot{r}$, we get 

$$e\cos \frac{c_0}{\sqrt{a}}=\frac{\vec{r}_0\cdot\vec{v}_0}{\sqrt{\mu a}}\tag{4.4-3}$$

Using the trig identity for the cosine of a sum we can write equation (4.3-5) as 

$$\sqrt{\mu}t=ax-ae\sqrt{a}\left(\cos\frac{x}{\sqrt{a}}\cos\frac{c_0}{\sqrt{a}}-\sin\frac{x}{\sqrt{a}}\sin\frac{c_0}{\sqrt{a}}-\cos\frac{c_0}{\sqrt{a}}\right).$$

Then substituting equations (4.4-1) and (4.4-3) and rearranging: 

>$$\sqrt{\mu}t=a(x-\sqrt{a}\sin\frac{x}{\sqrt{a}})+\frac{\vec{r}_0\cdot\vec{v}_0}{\sqrt{\mu}}a(1-\cos\frac{x}{\sqrt{a}})+r_0\sqrt{a}\sin\frac{x}{\sqrt{a}}\tag{4.4-4}$$


In a similar fashion we can use the trig identity for the sine of a sum 
to rewrite equation (4.3-4) as 

> $$r=a+ae(\sin\frac{x}{\sqrt{a}}\cos\frac{c_0}{\sqrt{a}}+\cos\frac{x}{\sqrt{a}}\sin\frac{c_0}{\sqrt{a}}).\tag{4.4-5}$$


Figure 4.4-1 The Kepler problem 

Substituting equations (4.4-1) and (4.4-3) we get: 

$$r=a+a\left[\frac{\vec{r}_0\cdot\vec{v}_0}{\sqrt{\mu a}}\sin\frac{x}{\sqrt{a}}+(\frac{r_0}{a}-1)\cos\frac{x}{\sqrt{a}}\right]\tag{4.4-6}$$

At this point let us introduce another new variable

$$z=\frac{x^2}{a}\tag{4.4-7}$$

Equation (4.4-4) then becomes 

$$\sqrt{\mu}t=\frac{x^2}{z}(x-\frac{x}{\sqrt{z}}\sin\sqrt{z})+\frac{\vec{r}_0\cdot\vec{v}_0}{\sqrt{\mu}}\frac{x^2}{z}(1-\cos\sqrt{z})+r_0\frac{x}{\sqrt{z}}\sin\sqrt{z}$$

which can be rearranged as 

$$\sqrt{\mu}t=\left[\frac{\sqrt{z}-\sin\sqrt{z}}{\sqrt{z^3}}\right]x^3+\frac{\vec{r}_0\cdot\vec{v}_0}{\sqrt{\mu}}x^2\frac{1-\cos\sqrt{z}}{z}+\frac{r_0 x\sin\sqrt{z}}{\sqrt{z}}\tag{4.4-8}$$

Similarly equation (4.4-6) becomes 

$$r=\frac{x^2}{z}+\frac{\vec{r}_0\cdot\vec{v}_0}{\sqrt{\mu}}\frac{x}{\sqrt{z}}\sin\sqrt{z}+r_0\cos\sqrt{z}-\frac{x^2}{z}\cos\sqrt{z}.\tag{4.4-9}$$

These equations are indeterminate for $z = 0$. To remedy this we will introduce two very useful functions which can be expressed as a series: 

>$$\begin{eqnarray*}
C(z)&\equiv&\frac{1-\cos\sqrt{z}}{z}=\frac{1-\cosh\sqrt{-z}}{z}=\frac{1}{2!}-\frac{z}{4!}+\frac{z^2}{6!}-\frac{z^3}{8!}+\dots \\
& &=\sum_{k=0}^\infty\frac{(-z)^k}{(2k+2)!}\tag{4.4-10}\\
S(z)&=&\frac{\sqrt{z}-\sin\sqrt{z}}{\sqrt{z^3}}=\frac{\sinh\sqrt{-z}-\sqrt{-z}}{\sqrt{(-z)^3}} \\
&=&\frac{1}{3!}-\frac{z}{5!}+\frac{z^2}{7!}-\frac{z^3}{9!}+\dots=\sum_{k=0}{\infty}\frac{(-z)^k}{(2k+3)!}.\tag{4.4-11}
\end{eqnarray*}$$

>*No derivation for the series forms are given, but I think you can get them from the Taylor series expansions of the terms of the trig forms. It must be the case that both forms evaluate to the same series expansion. When $z=0$, we just look at the first term of the series expansion of each function, so $C(0)=\frac{1}{2!}=\frac{1}{2}$ and $S(0)=\frac{1}{3!}=\frac{1}{6}$*

```python
def C(z):
    """
    Calculate equation 4.4-10. Use the closed form trig or hyperbolic trig functions
    when possible since they are essentially optimized series expansions themselves.
    Only use the first term of the series expansion when z=0.
    
    Use vectorized numpy trig or hyperbolic trig when possible.
    """
    def Cg(z):
        return (1-np.cos (np.sqrt( z)))/z
    def Cl(z):
        return (1-np.cosh(np.sqrt(-z)))/z
    def Cz(z):
        return 1/2
    result=z*0
    try:
        wg=np.where(z>0)
        wl=np.where(z<0)
        wz=np.where(z==0)
        result[wg]=Cg(z[wg])
        result[wl]=Cl(z[wl])
        result[wz]=Cz(z[wz])
    except:
        if z>0:
            result=Cg(z)
        elif z<0:
            result=Cl(z)
        else:
            result=Cz(z)
    return result

def S(z):
    """
    Calculate equation 4.4-11. See notes for C(z) above
    """
    def Sg(z):
        return ( np.sqrt( z)-np.sin (np.sqrt( z)))/np.sqrt(( z)**3)
    def Sl(z):
        return (-np.sqrt(-z)+np.sinh(np.sqrt(-z)))/np.sqrt((-z)**3)
    def Sz(z):
        return 1/6
    result=z*0
    try:
        wg=np.where(z>0)
        wl=np.where(z<0)
        wz=np.where(z==0)
        result[wg]=Sg(z[wg])
        result[wl]=Sl(z[wl])
        result[wz]=Sz(z[wz])
    except:
        if z>0:
            result=Sg(z)
        elif z<0:
            result=Sl(z)
        else:
            result=Sz(z)
    return result
    
```

The properties of these functions will be discussed in a later section. Using these functions equations (4.4-8) and (4.4-9) become 

>$$\sqrt{\mu}t=x^3 S+\frac{\vec{r}_0\cdot\vec{v}_0}{\sqrt{\mu}}x^2C+r_0x(1-zS).\tag{4.4-12}$$

>$$r=\sqrt{\mu}\frac{dt}{dx}=x^2 C+\frac{\vec{r}_0\cdot\vec{v}_0}{\sqrt{\mu}}x(1-zS)+r_0(1-zC).\tag{4.4-13}$$

## Solving for $x$ When Time is Known

An intermediate step to finding the radius and velocity vectors at a future time is to find $x$ when time is known. From $\vec{r}_0$, $\vec{v}_0$ and the energy equation you can obtain the semi-major axis, $a$. But how we have a problem; since equation (4.4-12) is transcendental in $x$, we cannot get it by itself on the left of the equal sign. Therefore, a trial-and-error solution is indicated. 

>*Initially it doesn't look bad -- It's cubic in $x$. But then we remember that $S$ and $C$ are functions of $z$ which itself is a function of $x$.*

Fortunately, the $t$ vs $x$ curve is well-behaved and a Newton iteration technique may be used successfully to solve for $x$ when time-of-flight is given. If we let $t_0=0$ and choose a trial value for $x$--call it $x_n$, then 

$$\sqrt{\mu}t_n=\frac{\vec{r}_0\cdot\vec{v}_0}{\sqrt{\mu}}x^2_nC+\left(1-\frac{r_0}{a}\right)x^3_nS+r_0 x_n\tag{4.4-14}$$ 

where $t_n$ is the time-of-flight corresponding to the given $\vec{r}_0$, $\vec{v}_0$, $a$ and the trial value of $x$. Equation (4.4-7) has been used to eliminate $z$. In one sense, $C$ and $S$ should have a subscript of $n$ also because they are functions of the guess of $x_n$.

```python
#All of this uses canonical units
#scalar components of r0v and v0v, which are perpendicular
r0=1
v0=1.4
h=r0*v0
r0dotv0=0 #since they are perpendicular
EE=v0**2/2-1/r0
a=-1/(2*EE)
print("a: %f"%a)
x=np.arange(-40,40,0.001)
z=x**2/a
t=x**3*S(z)+r0dotv0*x**2*C(z)+r0*x*(1-z*S(z))
plt.figure("4.4-2 left")
plt.plot(x,t)
plt.show()
```
